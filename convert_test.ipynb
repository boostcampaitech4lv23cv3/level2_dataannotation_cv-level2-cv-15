```python
import json
import re
```

```python
with open('/opt/ml/input/data/AIHUB/gt/textinthewild_data_info.json', 'rt', encoding='UTF-8') as F:
    data = json.load(F)
```

```python
data
```

```python
file_name_list = {}
```

```python
for image in data['images']:
    image_id = image['id']
    file_name = image['file_name']
    file_name_list[image_id] = file_name
```

```python
file_name_list
```

```python
print(len(file_name_list))
```

```python
data['annotations']
```

```python
bool(re.compile('^[a-zA-Z0-9]*$').sub('', 'a2'))
```

```python
ufo = {'images': {}}
for ann in data['annotations']:
    ann_id = ann['id']
    image_id = ann['image_id']
    text = ann['text']
    if text == None:
        pass
    elif bool(re.compile('^[a-zA-Z0-9]*$').sub('', text)) == False:
        lan = 'en'
    else:
        lan = 'ko'
    
    if image_id in file_name_list:
        file_name = file_name_list[image_id]
        x_min = ann['bbox'][0]
        y_min = ann['bbox'][1]
        x_max = x_min + ann['bbox'][2]
        y_max = y_min + ann['bbox'][3]
        
        #   
        points = [[x_min, y_min], [x_max, y_min], [x_max, y_max], [x_min, y_max]]
        

        if file_name in ufo['images']:
            ufo['images'][file_name]['words'][ann_id] = {'points': points}
            ufo['images'][file_name]['words'][ann_id]['illegibility'] = False
            ufo['images'][file_name]['words'][ann_id]['transcription'] = text
            ufo['images'][file_name]['words'][ann_id]['language'] = lan
        else:
            ufo['images'][file_name] = {'words': {ann_id: {'points': points}}}
            ufo['images'][file_name]['words'][ann_id]['illegibility'] = False
            ufo['images'][file_name]['words'][ann_id]['transcription'] = text
            ufo['images'][file_name]['words'][ann_id]['language'] = lan
```

```python
with open('/opt/ml/input/data/AIHUB/gt/new_text.json', 'w', encoding='UTF-8') as T:
    json.dump(ufo, T, indent=4)
```
